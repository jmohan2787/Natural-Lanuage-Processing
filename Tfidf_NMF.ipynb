```python
def SentilMentalAnalysis(data,columnname):
    
    import matplotlib.pyplot as plt
    import seaborn as sns
    from nltk.sentiment.vader import SentimentIntensityAnalyzer
    sid=SentimentIntensityAnalyzer()
    data.dropna(inplace=True)
    data['scores']=data[columnname].apply(lambda commentText: sid.polarity_scores(commentText))
    data['compound']=data['scores'].apply(lambda d:d['compound'])
    data['Negative']=data['scores'].apply(lambda d:d['neg'])
    data['Positive']=data['scores'].apply(lambda d:d['pos'])
    data['Neutral']=data['scores'].apply(lambda d:d['neu'])
    
    data['comp_score']  = data['compound'].apply(lambda c: 'pos'if c>0 else 'neg')
    # plt.cfg()
    
    posneg=pd.DataFrame(data['comp_score'].value_counts())
    return posneg,data
```

```python
posneg,data=SentilMentalAnalysis(df1,columnname='Chat')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer(max_df=0.95,min_df=2,stop_words='english')
dtm=tfidf.fit_transform(data['Chat'])
```

```python
from sklearn.decomposition import NMF
```

```python
from sklearn.decomposition import NMF
nfm_model=NMF(n_components=5,random_state=42)
nfm_model.fit(dtm)
```

```python
print(nfm_model)
```

```python
for topic in nfm_model.components_:
    results=[tfidf.get_feature_names_out()[i] for i in topic.argsort()[-10:]]
    print(results)
```

```python
topic_result=nfm_model.transform(dtm)
data['Topics']=topic_result.argmax(axis=1)
print(data)
```
